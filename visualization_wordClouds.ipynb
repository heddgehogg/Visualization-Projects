Word clouds (also known as tag clouds or text clouds) operate on a fairly simple algorithm: the more frequently specific words appear in the source of textual data (such as texts, speeches, blogs, etc.), the larger these words will be displayed in the word cloud.

For this task in Python, I used the appropriate library - `word_cloud`, developed by **Andreas Mueller**. You can read more about it at [this link](https://github.com/amueller/word_cloud/).

```python
%matplotlib inline

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') 
```

```python
# !pip install wordcloud

from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

Word clouds are used for high-level analysis and visualization of textual data. Here, I attempted to analyze *Alice's Adventures in Wonderland*, written by **Lewis Carroll**.

```python
STOPWORDS
```

```python
with open("data/alice_novel.txt", "r") as file:
    alice_novel = file.read()
```

```python
stopwords = set(STOPWORDS)
```

```python
alice_wc = WordCloud(
    background_color='white',
    max_words=2000,
    stopwords=stopwords
)

alice_wc.generate(alice_novel)
```

Great! The word cloud has been created, now it's time to visualize it.

```python
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Interesting! So, among the first 2000 words, the most common ones are **Alice**, **said**, **little**, **Queen**, etc. I increase the size of the word cloud so that you can better see less frequent words.

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Much better! However, the word "said" is not that informative. I add it to the list of stop words and regenerate the word cloud.

```python
stopwords.add('said') 
alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

Wonderful! And it looks intriguing! Another interesting feature of the `word_cloud` library is that I can insert the word cloud into an image of any shape. Let's use Alice and her rabbit's mask (*alice_mask.png*).

```python
import numpy as np
from PIL import Image
 
alice_mask = np.array(Image.open('images/alice_mask.png'))
    
print('Image downloaded and saved!')
```

```python
fig = plt.figure()
fig.set_figwidth(14) 
fig.set_figheight(18) 

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
alice_wc = WordCloud(background_color='white', colormap = 'Paired', max_words=2000, mask=alice_mask, stopwords=stopwords)

alice_wc.generate(alice_novel)

fig = plt.figure()
fig.set_figwidth(7)
fig.set_figheight(9) 

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
